# Create a namespace in the catalog

```python
%%sql

CREATE SCHEMA IF NOT EXISTS landing;
```

```python
%%sql
    
CREATE TABLE IF NOT EXISTS landing.testtable(
    id  int,
    category  varchar(20),
    amount numeric(16,2)
);
```
