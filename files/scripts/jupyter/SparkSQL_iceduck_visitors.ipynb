# SparkSQL Example

```python
Create a new Spark session for this notebook first. To identify it in the Spark UI at (http://localhost:8081) correctly the application will equal the notebooks filename.
```

```python
from pyspark.sql import SparkSession
import ipynbname

try:
    SparkSession.builder.master("spark://spark:7077").getOrCreate().stop()
except:
    print("Stop session failed")

try:    
    spark = SparkSession.builder \
        .appName(ipynbname.name()) \
        .master("spark://spark:7077") \
        .config("spark.sql.shuffle.partitions", "10") \
        .getOrCreate()
    print(spark)
except:
    print("Failed to create a new Spark Session" )

print("Spark session is ready! 🚀")
```

# Create a namespace in the catalog

Connection details for the Polaris Catalog are defined in the .pyiceberg.yaml file, which has been volumed into the spark container, so we can simply connect to the *default* catalog without specifiying connection details, credentials etc.

```python
%%sql

CREATE SCHEMA IF NOT EXISTS iceduck;
```

```python
%%sql
    
CREATE TABLE IF NOT EXISTS iceduck.visitors (
    visit_timestamp TIMESTAMP,
    visitor         VARCHAR(20),
    description     VARCHAR(50)
);
```

```python
%%sql
    
INSERT INTO iceduck.visitors VALUES ( now(), 'SparkSQL', 'created with the SparkSql in Jupyter' );
```

```python
%%sql
    
SELECT * FROM iceduck.visitors ORDER BY 1 DESC;
```
